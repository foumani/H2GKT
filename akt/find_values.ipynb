```python
import pandas as pd
import os

# Configuration
dataset_dir = "../data/assist2012"
csv_filename = "2012-2013-data-with-predictions-4-final.csv"

# Configuration
dataset_dir = "../data/assist2009"
csv_filename = "ASSISTments_2009_2010_skill_builder_data.csv"
```

```python
file_path = os.path.join(dataset_dir, csv_filename)

print(f"Reading {file_path}...")

# Read the CSV
# low_memory=False helps if the file is large and types are mixed
df = pd.read_csv(file_path,
                 encoding='ISO-8859-1',
                 low_memory=False)

# 1. SHOW THE COLUMNS AND DATA
print("\n--- CSV COLUMNS ---")
print(df.columns.tolist())

print("\n--- FIRST 5 ROWS ---")
print(df.head())

# 2. COUNT UNIQUE VALUES
print("\n--- STATISTICS ---")

# Check for Problem ID
if 'problem_id' in df.columns:
    n_pid = df['problem_id'].nunique()
    print(f"Unique 'problem_id' (n_pid): {n_pid}")
else:
    print("'problem_id' column not found.")

# Check for Skill ID (could be 'skill_id', 'sequence_id', or similar)
possible_skill_cols = ['skill_id', 'sequence_id', 'skill_name']
found_skill = False
for col in possible_skill_cols:
    if col in df.columns:
        n_skill = df[col].nunique()
        print(f"Unique '{col}' (n_question): {n_skill}")
        found_skill = True

if not found_skill:
    print("Could not automatically find a skill/sequence ID column. Check the column list above.")
```
